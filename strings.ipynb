Reverse string

```python
name = "Tushar"
print(name[::-1])
```

Palindrome string

```python
name = 'eye'
if name == name[::-1]:
    print('Palindrome string')
else:
    print('not a Palindrome string')
```

 Frequency of characters

```python
S = 'Tushar verma'

freq = {}

for x in S:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1

print(freq)
```

Remove vowels

```python
s = 'Tushar verma'
s_ = [] 
for x in s:
    if x not in 'a e i o u':
        s_.append(x)
print("".join(s_))
```

Count uppercase/lowercase

```python
s = 'Tushar VEA'
upper = 0
lower = 0
for x in s:
    if x.isupper():
        upper+=1
    elif x.islower():
        lower+=1

print( 'Ther is ' , upper)
print('There is '  , lower)
```
